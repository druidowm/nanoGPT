```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load wandb_export.csv
df = pd.read_csv('wandb_export.csv')
df.head()
```

```python
plt.plot(df["Step"], df["gpt2_last_char_to_char1729642919.4054031 - train/loss"], label="byte-to-byte")
plt.plot(df["Step"], df["gpt2_char_to_token1729491541.2756786 - train/loss"], label="byte-to-token")
plt.plot(df["Step"], df["gpt2 - train/loss"], label="token-to-token")
plt.plot(df["Step"], df["gpt2_last_char_token_to_char1729587337.8266432 - train/loss"], label="last-byte-token-to-byte")
plt.plot(df["Step"], df["gpt2_last_char_token_to_token1729493675.529459 - train/loss"], label="last-byte-token-to-token")
plt.legend()
plt.yscale('log')
plt.xlabel("Step")
plt.ylabel("Normalized Loss")
```

```python
plt.plot(df["Step"], df["gpt2_char_to_token1729491541.2756786 - train/loss"], label="byte-to-token")
plt.plot(df["Step"], df["gpt2_augmented_char_to_token1729630560.5729918 - train/loss"], label="augmented-byte-to-token")
plt.plot(df["Step"], df["gpt2_last_char_token_to_token1729493675.529459 - train/loss"], label="last-byte-token-to-token")
plt.legend()
plt.yscale('log')
plt.xlabel("Step")
plt.ylabel("Normalized Loss")
plt.show()
```
